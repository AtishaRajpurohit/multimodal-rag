```python
import numpy as np
import matplotlib.pyplot as plt

def generate_unit_vectors(n, d):
    """
    Generate n random unit vectors in d dimensions.
    """
    vecs = np.random.randn(n, d)  # Gaussian distribution
    vecs /= np.linalg.norm(vecs, axis=1, keepdims=True)  # Normalize to unit length
    return vecs

def cosine_sim(u, v):
    """
    Compute cosine similarity between a vector u and set of vectors v.
    """
    return v @ u / (np.linalg.norm(u) * np.linalg.norm(v, axis=1))

# Parameters
n = 10000
dimensions = [2, 10, 100, 1000]

plt.figure(figsize=(12, 8))

for i, d in enumerate(dimensions, 1):
    # Generate random unit vectors
    vecs = generate_unit_vectors(n, d)

    # Pick one random vector
    ref = vecs[np.random.randint(n)]

    # Compute cosine similarities with all others
    cosines = cosine_sim(ref, vecs)

    # Plot histogram
    plt.subplot(2, 2, i)
    plt.hist(cosines, bins=100, range=(-1, 1), color="skyblue", edgecolor="black")
    plt.title(f"d = {d}")
    plt.xlabel("Cosine similarity")
    plt.ylabel("Frequency")
    plt.xlim(-1, 1)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

def generate_unit_vectors(n, d):
    """
    Generate n random unit vectors in d dimensions.
    """
    vecs = np.random.randn(n, d)  # Gaussian distribution
    vecs /= np.linalg.norm(vecs, axis=1, keepdims=True)  # Normalize to unit length
    return vecs

def cosine_sim(u, v):
    """
    Compute cosine similarity between a vector u and set of vectors v.
    """
    return v @ u / (np.linalg.norm(u) * np.linalg.norm(v, axis=1))

# Parameters
n = 10000
dimensions = [2, 10, 100, 1000]

plt.figure(figsize=(12, 8))

for i, d in enumerate(dimensions, 1):
    # Generate random unit vectors
    vecs = generate_unit_vectors(n, d)

    # Pick one random vector
    ref = vecs[np.random.randint(n)]

    # Compute cosine similarities with all others
    cosines = cosine_sim(ref, vecs)

    # Calculate kurtosis (Fisher’s definition: normal dist = 0)
    k = kurtosis(cosines, fisher=True)

    # Plot histogram
    plt.subplot(2, 2, i)
    plt.hist(cosines, bins=100, range=(-1, 1), color="skyblue", edgecolor="black")
    plt.title(f"d = {d}")
    plt.xlabel("Cosine similarity")
    plt.ylabel("Frequency")
    plt.xlim(-1, 1)

    # Annotate kurtosis value
    plt.text(
        0.05, 0.95, f"Kurtosis: {k:.3f}", 
        transform=plt.gca().transAxes,
        fontsize=10, color="red",
        verticalalignment="top"
    )

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

def generate_unit_vectors(n, d):
    """
    Generate n random unit vectors in d dimensions.
    """
    vecs = np.random.randn(n, d)  # Gaussian distribution
    vecs /= np.linalg.norm(vecs, axis=1, keepdims=True)  # Normalize to unit length
    return vecs

def cosine_sim(u, v):
    """
    Compute cosine similarity between a vector u and set of vectors v.
    """
    return v @ u / (np.linalg.norm(u) * np.linalg.norm(v, axis=1))

# Parameters
n = 10000
dimensions = [2, 10, 100, 200, 250, 1000]  # added 200 and 250

plt.figure(figsize=(14, 10))

for i, d in enumerate(dimensions, 1):
    # Generate random unit vectors
    vecs = generate_unit_vectors(n, d)

    # Pick one random vector
    ref = vecs[np.random.randint(n)]

    # Compute cosine similarities with all others
    cosines = cosine_sim(ref, vecs)

    # Calculate kurtosis (Fisher: normal = 0)
    k = kurtosis(cosines, fisher=True)

    # Plot histogram
    plt.subplot(3, 2, i)
    plt.hist(cosines, bins=100, range=(-1, 1), color="skyblue", edgecolor="black")
    plt.title(f"d = {d}")
    plt.xlabel("Cosine similarity")
    plt.ylabel("Frequency")
    plt.xlim(-1, 1)

    # Annotate kurtosis
    plt.text(
        0.05, 0.95, f"Kurtosis: {k:.3f}", 
        transform=plt.gca().transAxes,
        fontsize=10, color="red",
        verticalalignment="top"
    )

plt.tight_layout()
plt.show()
```
